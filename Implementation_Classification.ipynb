# **Naive Bayes Implementation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Libraries for splitting the dataset
from sklearn.model_selection import train_test_split

#Libraries for feature preprocessing
from sklearn.preprocessing import StandardScaler

#Training
from sklearn.naive_bayes import GaussianNB

#Metrics
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(10)
```

```python
df.describe()
```

```python
# Age distribution
sns.histplot(x='Age', data=df)
```

```python
# Age and purchase distribution
sns.histplot(x='Age', hue='Purchased', data=df)
```

```python
# Purchase history distibution
sns.countplot(x='Purchased', data=df)
```

```python
X = df.iloc[:, :-1]
print(X)
y = df.iloc[:, -1]
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
# Feature Scaling

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Training - Naive Byes
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
# Predicting Results on Test Dataset
y_pred = nb.predict(X_test)
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```
